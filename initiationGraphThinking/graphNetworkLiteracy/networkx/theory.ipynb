```python
# small famous graphs

import pylab as plt

def plot_graph(g,name):
    pos=nx.spring_layout(g,iterations=100)
    plt.title(name)
    nx.draw(g,pos,node_size=50,with_labels=False)
```

```python
petersen = nx.petersen_graph()
plt.subplot(221)
plot_graph(petersen,'Petersen')
```

```python
tutte = nx.tutte_graph()
nx.draw(tutte)
```

```python
maze = nx.sedgewick_maze_graph()
nx.draw(maze)
```

```python
tet = nx.tetrahedral_graph()
nx.draw(tet)
```

```python
# classic graphs
```

```python
K_5 = nx.complete_graph(5)
nx.draw(K_5)
```

```python
K_3_5 = nx.complete_bipartite_graph(3, 5)
nx.draw(K_3_5)
```

```python
barbell = nx.barbell_graph(10, 10)
nx.draw(barbell)
```

```python
lollipop = nx.lollipop_graph(10, 20)
nx.draw(lollipop)
```

```python
# random graphs
```

```python
er = nx.erdos_renyi_graph(100, 0.15)

plt.subplot(221)
plot_graph(er,'Erdos-Renyi')
```

```python
ws = nx.watts_strogatz_graph(30, 3, 0.1)
plt.subplot(223)
plot_graph(ws,'Watts-Strogatz')
```

```python
ba = nx.barabasi_albert_graph(100, 5)
plt.subplot(222)
plot_graph(ba,'Barabasi-Albert')
```

```python
lob = nx.random_lobster(8, 0.9, 0.9)

plt.subplot(224)
plot_graph(lob,'Random lobster')

plt.savefig('./data_out/random_graphs.pdf')
```
